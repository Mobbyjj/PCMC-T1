```python
import matplotlib.pyplot as plt
import numpy as np
import os
rootdir = '/media/ssd/fanwen/T1mapping/Free_breathing/Model'
```

```python
models = ['ISMRM','MICCAI_no_dice', 'MICCAI']
segs = []
tests = []

segs.append(np.load(os.path.join('/media/ssd/fanwen/T1mapping/Free_breathing/full_set_segmentation.npy'))[180:])
tests.append(np.load(os.path.join('/media/ssd/fanwen/T1mapping/Free_breathing/full_set_data_normed.npy'))[180:])
for model in models:
    # load registered_seg_set_final30.npy; registered_test_set_final30.npy
    seg = np.load(os.path.join(rootdir, model, 'registered_seg_set_final30.npy'))
    test = np.load(os.path.join(rootdir, model, 'registered_test_set_final30.npy'))
    segs.append(seg)
    tests.append(test)
```

```python
segs = np.stack(segs, axis=0)
tests = np.stack(tests, axis=0)
```

```python
print(segs.shape)
```

```python
segs.shape
ncase = 11
nslice = 2
# 4 * 11
fig, axs = plt.subplots(4, 11, figsize=(11, 4))
for i in range(11):
    for j in range(4):
        axs[j,i].imshow(segs[j,ncase,...,i,nslice], cmap='gray')
        axs[j,i].axis('off')
```

```python
fig, axs = plt.subplots(4, 11, figsize=(11, 4))
for i in range(11):
    for j in range(4):
        axs[j,i].imshow(tests[j,ncase,...,i,nslice], cmap='gray')
        axs[j,i].axis('off')
```
